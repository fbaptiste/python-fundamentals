Let's load some data up using Pandas:

````python
df = pd.read_csv('DEXUSEU.csv')
df.iloc[:5, :]
````

````python
df.info()
````

````python
df.describe(include='all')
````

Let's first handle the `DATE` column - these were loaded as string:

````python
df.iloc[0, 0], type(df.iloc[0, 0])
````

First thing we'll do, is convert that to actual datetimes:

````python
pd.to_datetime(df['DATE'], errors='coerce')
````

Then, we also have to convert the exchange rates to floats - you'll remember that we have some `.` points in the data - but this will actually work well for us - when we coerce the data to floats, the `.` will result in `NaN`:

````python
rates = pd.to_numeric(df['DEXUSEU'], errors='coerce')
rates
````

Now we have `NaN` data in our exchange rates:

````python
rates[rates.isnull()]
````

We could drop the rows with `NaN`, but here I'll choose to use linear interpolation to fill the missing values:

````python
rates.interpolate(method='linear')
````

Let's build our new data set:

````python
data = pd.concat(
    [
        pd.to_datetime(df['DATE'], errors='coerce'),
        pd.to_numeric(
            df['DEXUSEU'], errors='coerce'
        ).interpolate(method='linear')
    ],
    axis=1
)
data[:5]
````

````python
data.info()
````

So now we want to chart the exchange rate. Matplotlib does not always work well with Pandas series objects, so we should convert our series to regular NumPy arrays before we plot the data (we use the `.values` attribute):

````python
data['DATE'].values
````

````python
data['DEXUSEU'].values
````

````python
fig, ax = plt.subplots()
ax.plot(data['DATE'].values, data['DEXUSEU'].values, label='rate')
ax.set_xlabel('date')
ax.set_ylabel('exchange rate')
ax.set_title('US/EU Exchange Rate')
````

Now it might be nice to add a moving average to the data abd plot that on the same chart.

Pandas has a `.rolling()` method that essentially groups data in a rolling window, and we can then calculate the mean in each window:

````python
m5_avg = data['DEXUSEU'].rolling(window=5).mean()
m5_avg
````

````python
fig, ax = plt.subplots()
ax.plot(
    data['DATE'].values, 
    data['DEXUSEU'].values, 
    label='rate'
)
ax.plot(
    data['DATE'].values, 
    data['DEXUSEU'].rolling(window=50).mean(), 
    label='ma'
)
ax.set_xlabel('date')
ax.set_ylabel('exchange rate')
ax.legend()
ax.set_title('US/EU Exchange Rate')
````

(if you're wondering why we're seeing that `Text(...)` line above our chart, that's Jupyter printing out the last result (which was setting the title):

````python
ax.set_title('US/EU Exchange Rate')
````

Next, let's look at how we might use NumPy's `histogram` function, to plot a frequency distribution of some data.

Let's generate random data using a Gaussian distribution.

We saw how to do this using Python's random module:

````python
import random
````

````python
random.seed(0)
data_1 = [random.gauss(0, 1) for _ in range(1_000_000)]
````

We could also just use the NumPy random module:

````python
np.random.seed(0)
data_2 = np.random.normal(0, 1, size=1_000_000)
````

What we now want to do is create a frequency distribution chart, and see if it looks like a normal distribution:

````python
freq_1, bins_1 = np.histogram(data_1, bins=50)
freq_2, bins_2 = np.histogram(data_2, bins=50)
````

Remember that the output of `np.histogram` is a **tuple** containing the frequencies and the bins.

Remember also that the bins contain the righmost edge in addition to the left edges - so we'll omit this when charting.

Now let's plot each one and see what the distribution looks like, using a bar chart as well:

````python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.bar(
    bins_1[:-1], 
    freq_1
)
ax1.set_title('Python Generated')

ax2.bar(
    bins_2[:-1], 
    freq_2
)
ax2.set_title('NumPy Generated')
````

In this case, you may prefer a scatter plot instead of a bar chart:

````python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.scatter(
    bins_1[:-1], 
    freq_1
)
ax1.set_title('Python Generated')

ax2.scatter(
    bins_2[:-1], 
    freq_2
)
ax2.set_title('NumPy Generated')
````

````python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(
    bins_1[:-1], 
    freq_1
)
ax1.set_title('Python Generated')

ax2.plot(
    bins_2[:-1], 
    freq_2
)
ax2.set_title('NumPy Generated')
````

We may want to draw some vertical lines on the chart to indicate the 10th, 50th and 90th percentiles:

````python
percentiles_1 = [np.percentile(data_1, p) for p in (10, 50, 90)] 
percentiles_2 = [np.percentile(data_2, p) for p in (10, 50, 90)] 
````

````python
percentiles_1, percentiles_2
````

We can use the `axvline` method to do so:

````python
fig, ax = plt.subplots()
ax.scatter(
    bins_1[:-1], 
    freq_1,
    label='random'
)
ax.axvline(percentiles_1[0], linestyle=':', color='green', label='10%')
ax.axvline(percentiles_1[1], linestyle='--', color='red', label='50%')
ax.axvline(
    percentiles_1[2], 
    linestyle=':', 
    color=mpl.colors.to_rgba((0.1, 0.9, 0.1)), label='90%')
ax.legend()
````

Just like NumPy has a histogram function, MatPlotLib also has such a function which we can use directly, instead of using NumPy first.

````python
fig, ax = plt.subplots()
_ = ax.hist(data_2, bins=50)
````

(Note that I used `_ = ` because I didn't want to see the output of the `hist` method):

````python
ax.hist(data_2, bins=10)
````

which as you can see actually returns the frequencies and bins.

````python
import mplfinance as mpf
````

````python
df = pd.read_csv('AAPL_data.csv')
df.info()
````

Since all this data is for `AAPL`, we don't actually need that `Name` column.

In addition, I want to make the dates the index for the rows - I could load the data first, then convert to datetime and then set the index, but we can actually do this as we load the data from csv.

Lastly, we need the columns to be in the order Open/High/Low/Close/Volume - makes it easier to work with `mpf`. Fortunately that's the case in our csv file (otherwise we could just use fancy indexing to pick the columns in the required order).

````python
data = pd.read_csv(
    'AAPL_data.csv', 
    usecols=['date', 'open', 'high', 'low', 'close', 'volume'],
    index_col=0,
    parse_dates=[0],  # instructs Pandas to parse column 0 as dates
    skipinitialspace=True
)
data.info()
````

````python
data
````

Given that our index is a DateTime index, and our columns are in the correct order Open/High/Low/Close/Volume, using `mpf` to plot OHLC charts is quite easy.

````python
mpf.plot(data)
````

There's quite a bit of data to chart, and it is too compressed, so I'll plot only a subset of the data, and make the chart a bit larger:

````python
mpf.plot(data.iloc[:150], figsize=(18, 8))
````

We can change the style of the chart:

````python
mpf.plot(data.iloc[:150], type='candle', figsize=(18, 8))
````

We can even add moving average lines very easily:

````python
mpf.plot(data.iloc[:150], type='ohlc', mav=(5, 30), figsize=(18, 8))
````

We can also add the Volume as a second subplot, and show the non-trading day gaps:

````python
mpf.plot(
    data.iloc[:150], 
    type='ohlc', 
    mav=(5, 30), 
    volume=True, 
    show_nontrading=True, 
    figsize=(18, 8)
)
````

We can further add customizations to the chart, such as titles, axis labels, etc:

````python
mpf.plot(
    data.iloc[:150], 
    type='ohlc', 
    mav=(5, 30), 
    volume=True, 
    show_nontrading=True, 
    title='AAPL Daily',
    ylabel='OHLC',
    ylabel_lower='Volume',
    figsize=(18, 8)
)
````

We could also add additional plots, such as Bollinger bands (using teh standard period of 20 and 2 std deviations).

````python
period = 20
std = 2
data['ma20'] = data['close'].rolling(window=period).mean()
data['std20'] = data['close'].rolling(window=period).std()
data['upper'] = data['ma20'] + data['std20'] * std
data['lower'] = data['ma20'] - data['std20'] * std

data
````

We first create the additional subplots using just the columns we want (and the same rows as we'll use in the main plot):

````python
data[['upper', 'ma20', 'lower']].iloc[:150]
````

````python
bollinger = mpf.make_addplot(
    data[['upper', 'ma20', 'lower']].iloc[:150]
)
````

Then we add them to the main plot using the `addplot` argument:

````python
mpf.plot(
    data.iloc[:150], 
    addplot=bollinger,
    type='ohlc', 
    volume=True, 
    show_nontrading=True, 
    title='AAPL Daily',
    ylabel='OHLC',
    ylabel_lower='Volume',
    figsize=(18, 8)
)
````
